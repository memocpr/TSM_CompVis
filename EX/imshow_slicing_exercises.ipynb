 Imports and inline plotting

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

Cell 2 — Create test image (100x200 uint8 gradient)

```python
# Create a 100x200 uint8 test image with a horizontal gradient (0..255).
# This makes intensity changes obvious when we tweak vmin/vmax.
x = np.linspace(0, 255, 200, dtype=np.uint8)
im = np.tile(x, (100, 1))
im.shape, im.dtype
```

 Baseline display (vmin=0, vmax=255)

```python
plt.figure(figsize=(6, 3))
plt.imshow(im, vmin=0, vmax=255, cmap='gray')
plt.title("imshow(im, vmin=0, vmax=255, cmap='gray')")
plt.colorbar(orientation='horizontal');
```

Cell 4 — vmin=0, vmax=100

```python
plt.figure(figsize=(6, 3))
plt.imshow(im, vmin=0, vmax=100, cmap='gray')
plt.title('vmin=0, vmax=100')
plt.colorbar(orientation='horizontal');
```

Cell 5 — vmin=100, vmax=255


```python
plt.figure(figsize=(6, 3))
plt.imshow(im, vmin=100, vmax=255, cmap='gray')
plt.title('vmin=100, vmax=255')
plt.colorbar(orientation='horizontal');
```

Cell 6 — vmin=100, vmax=155



```python
plt.figure(figsize=(6, 3))
plt.imshow(im, vmin=100, vmax=155, cmap='gray')
plt.title('vmin=100, vmax=155')
plt.colorbar(orientation='horizontal');
```

Cell 7 — Invert image: 255 - im


```python
plt.figure(figsize=(6, 3))
plt.imshow(255 - im, vmin=0, vmax=255, cmap='gray')
plt.title('imshow(255 - im, vmin=0, vmax=255)')
plt.colorbar(orientation='horizontal');
```
